# rastrigin function with global search

```python
import numpy
import porgo
```

```python
rastrigin_function = lambda x: 10 * len(x) + numpy.sum(x**2 - 10*numpy.cos(2*numpy.pi*x)) # non-convex function
```

```python
bounds = [(-10, 10)] * 10 # high-dimensional search interval
```

```python
rastrigin = porgo.glos(rastrigin_function, bounds) # mutation=0.5, recombination=0.9
```

```python
rastrigin.rand_pop(100, True) # init population
```

```python
rastrigin.train_gen(3000) # if execute 1 time is not enough to converged, run this shell again.
# rastrigin.train_gen(2000) # run this command 2 times
# rastrigin.train_gen(1000) # run this command 3-4 times
```

```python
rastrigin.result()
rastrigin.mini, rastrigin.fit_mini # the best converged result
# print('minimum {}, medium {}, maximum {}'.format(rastrigin.fit_mini, rastrigin.fit_medi, rastrigin.fit_maxi))
```
