# multimodal function with global search

```python
import numpy
import porgo
```

```python
multimodal_function = lambda x: numpy.sum(x**2 - 10 * numpy.cos(2 * numpy.pi * x)) + 10 * len(x)
```

```python
bounds = [(-10, 10)] * 10 # high-dimensional search interval
```

```python
multimodal = porgo.glos(multimodal_function, bounds) # mutation=0.5, recombination=0.9
```

```python
multimodal.rand_pop(100, True) # init population
```

```python
multimodal.train_gen(3000) # if execute 1 time is not enough to converged, run this shell again.
# multimodal.train_gen(1000) # run this command 3 times
# multimodal.train_gen(2000) # run this command 2 times
```

```python
multimodal.result(minimum=True)
# print('minimum {}, medium {}, maximum {}'.format(multimodal.fit_mini, multimodal.fit_medi, multimodal.fit_maxi))
```
